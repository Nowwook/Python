# 자료형 변환
오브젝트 > 숫자 등등

```python
a= pd.to_datetime(dd['Died'],format='%Y-%m-%d')  
b= pd.to_datetime(dd['Born'],format='%Y-%m-%d')
a
```

```python
# 프레임에 적용
dd['born_dt'],dd['died_dt']=(b,a)
```

```python
dd
```

```python
dd['age_days']=(dd['died_dt']-dd['born_dt'])
dd
```

```python
dd['age_days']/365
```

```python
dd['born_dt'].dt.year
```

```python
# 데이터 섞기
dd.sample(frac=0.5)  # 랜덤으로 1/2만 들고오기
```

```python
dd.sample(frac=1)
```
