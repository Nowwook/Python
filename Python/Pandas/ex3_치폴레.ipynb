### Step 5. item_price에서 $를 없애고 float 형식으로 새롭게 저장해주세요.

```python
def re(x):
    x=x.replace('$','')
    return float(x)
        
chipo['item_price'].apply(re)
```

### Step 10. 한 주문에 20달러 이상 지불한 손님을 출력해주세요.

```python
t=chipo[chipo['item_price']>=20]
t
```

```python
d=chipo.groupby('order_id')['item_price'].sum()
c=d[d>=20].index
```

```python
e.groupby('item_name').count()
```

```python
f= chipo.set_index('order_id').loc[c]
f.index.unique()
```
