# 열이름 관리

```python
ebola=pd.read_csv('data/country_timeseries.csv')
ebola
```

```python
ebola.columns
```

```python
ebola.iloc[:5,[0,1,2,3,10,11]]
```

```python
long = pd.melt(ebola,id_vars=['Date','Day'])
long.head()
```

```python
# str 은 접근자
split = long.variable.str.split('_')
split
```

```python
# 분리된 데이터 확인
status = split.str.get(0)
status
```

```python
# 분리된 데이터를 합치기 concat
sp =long.variable.str.split('_',expand=True)
sp.columns =['status','country']
pd.concat([long,sp],axis=1)
```

# 중복 데이터 처리

```python
billboard = pd.read_csv('data/billboard.csv')
billboard
```

```python
bill_long = pd.melt(billboard,
                   id_vars=['year','artist','track','time','date.entered'],
                   var_name='week',value_name='rating')
print(bill_long.shape)
bill_long
```

```python
# 제목으로 중복데이터 현황 확인
bill_long[bill_long.track == 'Loser']
```

```python
song = bill_long[['year','artist','track','time']]
print(song.shape)
song
```

```python
# 중복 제거
songs = song.drop_duplicates()
songs
```

```python
# 순번 추가
songs['id'] = range(len(songs))
```

```python
rating = bill_long.merge(songs,
                        on=['year','artist','track','time'])
rating
```

```python
import os
import urllib.request
```

```python
#
txt 깃허브 url 로 데이터 가져오기
```

```python
with open('data/raw_data_urls.txt','r') as data_urls:
    for line,url in enumerate(data_urls):
        if line == 5:
            break
        fn =url.split('/')[-1].strip()   # 파일명 추출
        fp = os.path.join('','data',fn)  #폴더 저장 경로 지정
        print(url)
        print(fp)
        urllib.request.urlretrieve(url,fp)
```
